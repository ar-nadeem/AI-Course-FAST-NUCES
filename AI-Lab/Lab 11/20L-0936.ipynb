```python
import pandas as pd
import numpy as np
```

```python
ttrain = pd.read_csv('titanic_train.csv')
ttest = pd.read_csv('titanic_test.csv')
```

<h1 align="center">Preprocessing</h1>

# Convert the following into numerical values
* Sex (Male = 1 Female = 0)
* Embarked (S = 1 C = 2 Q = 3 N/A = 0)

# Drop the following
* Ticket
* Cabin

```python
for i,val in enumerate(ttrain.Sex):
    if val == 'male':
        ttrain.Sex[i] = 1
    elif val == 'female':
        ttrain.Sex[i] = 0


for i,val in enumerate(ttrain.Embarked):
    if val == 'S':
        ttrain.Embarked[i] = 1
    elif val == 'C':
        ttrain.Embarked[i] = 2
    elif val == 'Q':
        ttrain.Embarked[i] = 3
        

ttrain.head()
```

```python
ttrain.drop('Cabin',axis=1,inplace=True)
ttrain.drop('Ticket',axis=1,inplace=True)
```

```python
ttrain.dropna(inplace=True)
ttrain.head()
```

```python
# Passenger ID and Name are not required - They are just identifier and have no effect on the survival
x = ttrain[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
y = ttrain['Survived']
x=np.nan_to_num(x)
y=np.nan_to_num(y)
```

```python
#https://medium.com/analytics-vidhya/how-to-implement-multiple-linear-regression-from-scratch-c7ea23304540

def compute_cost(X, y, theta):
    m = len(y)
    y_pred = np.dot(X, theta)
    cost = (1/(2*m)) * np.sum((y_pred - y)**2)
    return cost

# Define the gradient descent function
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    cost_history = np.zeros(iterations)
    
    for i in range(iterations):
        y_pred = np.dot(X, theta)
        theta = theta - (alpha/m) * np.dot(X.T, (y_pred - y))
        cost_history[i] = compute_cost(X, y, theta)
        if (i % 2 == 0):
            print (cost_history[i])
        
    return theta, cost_history






# Initialize theta to zeros
theta = np.zeros((x.shape[1], 1))

# Set the hyperparameters
alpha = 0.01
iterations = 20

# Run gradient descent on the training data
theta, cost_history = gradient_descent(x, y.reshape(-1, 1), theta, alpha, iterations)



y_pred = np.dot(x, theta)
mse = np.mean((y.reshape(-1, 1) - y_pred)**2)
rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)
```

```python
# Evaluate the performance on the test data
ttest.dropna(inplace=True)
testx=ttest[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
testx=testx.apply(pd.to_numeric, errors='coerce')

testx=np.nan_to_num(testx)

y_pred = np.dot(testx, theta)
for vals in y_pred:
    print(vals)
    
```
