Question 1 | Genetic Algoritm

```python
# Genetic Algo Class
import random
from bootifulPrint import bootifulPrint
p = bootifulPrint()
class Genetic:
    def __init__(self, solutionSpace = 100, generationSpace = 5, bestSolutionSpace = 10):

        self.solutionSpace = solutionSpace # Total number of random solutions to generate
        self.generationSpace = generationSpace # Total number of generations to run after mutation
        self.bestSolutionSpace = bestSolutionSpace # Total space of best solution to be extracted from the current generation for mutation

        self.currentGen = [] # Current Generation that will be used to create the next generation
        self.__generateRandomSolution = None # Function to generate random solution
        self.__fitness = None # Function to calculate fitness of a solution
        self.__mutate = None # Mutation Function
        self.__cross = None # Cross Function
        self.debug = False # Debug mode
        self.globalBest = [] # Global best solutions

    # Set the debug mode
    def setDebug(self,b):
        self.debug = b

    # Set the function to generate random solution
    def setSolution(self, func):
        self.__generateRandomSolution = func
    # Set the function for fitness calculation
    def setFitness(self, func):
        self.__fitness = func

    # Set mutation function
    def setMutate(self, func):
        self.__mutate = func

    # Set Crossing function
    def setCross(self, func):
        self.__cross = func
    # Generate a 2d Array board for bootiful printing
    def printBoard(self,val):

        myarr = [[0 for i in range(8)] for i in range(8)]
        for i in range(8):
            myarr[val[i]][i] = 1

        for arr in myarr:
            p.printArray(arr)

    # Print information about the cur|ent generation
    def __print(self,s):
        # Print the best 3 solutions
        if self.debug:
            print("\n===",s," Generation===")
            print(f"Best Solution 0 = {self.bestSoltuions[0][0]} With Fitness ={self.bestSoltuions[0][1]}")
            print(f"Best Solution 1 = {self.bestSoltuions[1][0]} With Fitness ={self.bestSoltuions[1][1]}")
            print(f"Best Solution 2 = {self.bestSoltuions[2][0]} With Fitness ={self.bestSoltuions[2][1]}")

    # Generate the first generation | Random Solutions + Fitness
    def __genesisGeneration(self):
        if (self.__generateRandomSolution == None or self.__fitness == None or self.__mutate == None or self.__cross == None):
            print ("\n Please set solution, fitness, cross and mutation functions \n")
            exit(1)
        

        # Append the random solution and its fitness to the current generation
        for i in range(self.solutionSpace):
            sol = self.__generateRandomSolution()
            self.currentGen.append((sol,self.__fitness(sol)))
            
        # Sort the current generation based on fitness
        self.currentGen.sort(key = lambda x: x[1], reverse = False)

        # Set bestSolutions to the best solutions from the current generation
        self.bestSoltuions = self.currentGen[:self.bestSolutionSpace]



    # Mutate the best solutions to create the next generation
    def __nextGeneration(self):
        # Empty the current generation
        self.currentGen = []

        # Mutate the best solutions and append them to the current generation
        # Pick random solution from bestSolution and apply mutation to generate a new generation of solutions
        for i in range(self.solutionSpace):
            sol = self.__cross(random.choice(self.bestSoltuions)[0],random.choice(self.bestSoltuions)[0]) # Chosse random to cross
            sol = self.__mutate(sol) # Mutate
            self.currentGen.append((sol,self.__fitness(sol))) # Append Fitness
        
        # Empty bestSolutions
        self.bestSoltuions = []

        # Sort the current generation based on fitness
        self.currentGen.sort(key = lambda x: x[1], reverse = False)

        # Set bestSolutions to the best solutions from the current generation
        self.bestSoltuions = self.currentGen[:self.bestSolutionSpace]


    
    
    # Run the genetic algorithm
    def run(self):
        # Generate the First generation
        self.__genesisGeneration()
        self.__print("Genesis")

        self.globalBest.append(self.bestSoltuions[0])
        # Check if we reached the goal state of 0 fiteness
        if (self.bestSoltuions[0][1] == 0):
            self.printBoard(self.bestSoltuions[0][0])
            return self.bestSoltuions[0]

        for i in range(self.generationSpace):
            # Generate the next generation
            self.__nextGeneration() # Create crossed and mutated solutions and changes the current generation
            self.__print(f"{i}") # Prints current Gen ifo
            self.globalBest.append(self.bestSoltuions[0])

            # Check if we reached the goal state of 0 fiteness
            if (self.bestSoltuions[0][1] == 0):
                self.printBoard(self.bestSoltuions[0][0])
                return self.bestSoltuions[0]

        self.globalBest.sort(key = lambda x: x[1], reverse = False)

        # Return the best solution
        print("\n===BEST SOLUTION=== FITNESS ",self.globalBest[0][1],"\n")
        print(self.globalBest[0][0])
        self.printBoard(self.globalBest[0][0])
        return self.globalBest
        

```

Question 2 | 8 Queen problem with genetic algo

```python
# Genetic Algorithm
# Parameters are Solution Space, Generations, Best Solution
g = Genetic(1000,500,300)  
g.setDebug(True)

# 8 Queens Problem | Tuple of 8 numbers from 0 to 7
def solution():
    return tuple(random.randint(0,7) for i in range(8))

# Move queen one step in a random direction (Up and Down) with wrap around | If queen is at 0 and direction is -1, it will move to 7
def mutation(solution):
    # Convert the tuple to list so that we can mutate it
    sol = list(solution)

    for val in sol:
        if  random.randint(0,1) == 1: # 50% chance of mutation | But if we get 1, we will mutate
            val = (val + random.randint(-1,1)) % 8

    # Convert back to tuple and return
    return tuple(sol)


def cross(a, b):
    # Pick a random index to cross
    index = random.randint(0,7)
    # Create a new solution
    newSol = a[:index] + b[index:]
    return newSol

# Fitness is the number of queens that are not attacking each other
# [0,1,2,3,4,5,6,7]
def fitness(solution):
    count = 0
    # Convert to 2D array
    myarr = [[0 for i in range(8)] for i in range(8)]
    for i in range(8):
        myarr[solution[i]][i] = 1


    for i in range(8):
        for j in range(8):
            if myarr[i][j] == 1:  # found a queen
                # check horizontal and vertical attacks
                for k in range(8):
                    if myarr[i][k] == 1 and k != j:
                        count += 1
                    if myarr[k][j] == 1 and k != i:
                        count += 1
                # check diagonal attacks
                for k in range(1, 8):
                    if i-k >= 0 and j-k >= 0 and myarr[i-k][j-k] == 1:
                        count += 1
                    if i-k >= 0 and j+k < 8 and myarr[i-k][j+k] == 1:
                        count += 1
                    if i+k < 8 and j-k >= 0 and myarr[i+k][j-k] == 1:
                        count += 1
                    if i+k < 8 and j+k < 8 and myarr[i+k][j+k] == 1:
                        count += 1
    return count


g.setSolution(solution)
g.setFitness(fitness)
g.setMutate(mutation)
g.setCross(cross)
#g.printBoard((0,1,2,3,4,5,6,7))

g.run()
```
