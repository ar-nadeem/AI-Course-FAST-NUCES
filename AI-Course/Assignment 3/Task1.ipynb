```python
#Min-Max Algorithm
from math import inf as infinity
from random import choice
import platform
import numpy as np
from os import system
```

```python
Human = -1
Computer = +1
draw = 0
```

```python
#Initialize empty board
board = np.zeros((3,3),dtype=np.int32)
board
```

```python
def evaluate(curr_state):
    #write your code here to assign score if computer or human wins and 0 in case of draw; return score 
    if wins(curr_state, Human):
        score = +1
    elif wins(curr_state, Computer):
        score = -1
    else:
        score = 0

    return score


def wins(state, player):
    #return True or False by checking all possibilites of win for human or computer
    for x in range(3):
        if state[x,0] == state[x,1] == state[x,2] == player: # Row
            return True
    for y in range(3):
        if state[0,y] == state[1,y] == state[2,y] == player: # Column
            return True
    if state[0,0] == state[1,1] == state[2,2] == player: # Right Diagonal
        return True
    if state[0,2] == state[1,1] == state[2,0] == player: # Left Diagonal
        return True
    return False


def game_over(curr_state):
    # return True or False if human or computer wins
    if wins(curr_state, Computer) or wins(curr_state, Human):
        return True
    else:
        return False

    
    


def empty_cells(curr_state):
    # create a list of all empty cells[x,y] and return emptyCells list
    emptyCells = []
    for x in range(3):
        for y in range(3):
            if curr_state[x,y] == 0:
                emptyCells.append([x,y])
    return emptyCells


def valid_move(x, y):
    # check valid move if x,y is in emptyList return true else return false
    if board[x,y] == 0:
        return True
    else:
        return False
    

def set_move(x, y, player):
    # check selected cell[x,y]; if it is valid move then place player move on the board
    if valid_move(x,y):
        board[x,y] = player
        return True
    else:
        return False
    
def minimax(curr_state, depth, player):
    #Implement min-max algorithm here with following inputs:
    # Input:current-state of board, position of state where depth=[0-9], also player:human or computer
    #Output: list =[best row,best column,best score]
   
    if player == Computer:
        best = [-1, -1, -infinity]
    else:
        best = [-1, -1, +infinity]

    if depth == 0 or game_over(curr_state):
        score = evaluate(curr_state)
        return [-1, -1, score]
    
    for cell in empty_cells(curr_state):
        x, y = cell[0], cell[1]
        curr_state[x, y] = player
        score = minimax(curr_state, depth - 1, -player)
        curr_state[x, y] = 0
        score[0], score[1] = x, y

        if player == Computer:
            if score[2] > best[2]:
                best = score
        else:
            if score[2] < best[2]:
                best = score

    return best


    

def clean():
    os_name = platform.system().lower()
    if 'windows' in os_name:
        system('cls')
    else:
        system('clear')


def printBoard(curr_state, comp_choice, human_choice):
    #print board e.g.
    # ---------------
    # |   ||   ||   |
    # ---------------
    # |   ||   ||   |
    # ---------------
    # |   ||   ||   |
    # ---------------

    chars = {
        -1: human_choice,
        +1: comp_choice,
        0: ' '
    }
    str_line = '---------------'

    print('\n' + str_line)
    for row in curr_state:
        for cell in row:
            symbol = chars[cell]
            print(f'| {symbol} |', end='')
        print('\n' + str_line)


def computer_turn(comp_choice, human_choice):
    # comp_choice: X or 0, human_choice: X or 0 for depth <9
    depth = len(empty_cells(board))
    
    # Validation Check
    if depth == 0 or game_over(board):
        return
    else:
        clean()
        printBoard(board, comp_choice, human_choice)

    


def human_turn(comp_choice, human_choice):
    # comp_choice: X or 0, human_choice: X or 0 for depth <9
    depth = len(empty_cells(board))

    if depth == 0 or game_over(board):
        return
    else:
        clean()
        #print(f'Computer turn [{comp_choice}]')
        c = tuple(input("Enter your move (x,y): ").split(","))
        set_move(int(c[0]), int(c[1]), Human)
        printBoard(board, comp_choice, human_choice)
```

```python

def main():
    
    clean()
    h_choice = ''  # X or O
    c_choice = ''  # X or O
    first = ''  # if human is the first

    # Human chooses X or O to play
    h_choice = input('Choose X or O\nChosen: ').upper()
        
    # Setting computer's choice
    if h_choice == 'X':
        c_choice = 'O'
    else:
        c_choice = 'X'

    # Ask human want to start first: Y or N
    clean()
    
    first = input('First to start?[y/n]: ').upper()

    #Start Working to call other functions here 
    if (first == "Y"):
        human_turn(c_choice, h_choice)
    else:
        computer_turn(c_choice, h_choice)

    
    exit()


```

```python
#Call main function here
if __name__ == '__main__':
    main()
```
